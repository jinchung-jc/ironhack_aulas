## Lets get all columns where the displacement is missing.

```python
# passo a passo
vehicles['displ']
```

```python
# get a mask of which rows are missing
vehicles['displ'].isnull().sum()
```

```python
# create a new dataframe for those rows where the displacement is missing.
```

```python
condition = vehicles['displ'].isnull()
vehicles.loc[condition, :]
```

```python
vehicles.loc[condition, ['year', 'make', 'model', 'trany', 'drive','fuelType','cylinders', 'displ']]
```

```python
## There's some explanation for why the displacement is zero. What is it?
```

```python
vehicles.loc[condition, ['fuelType','displ', 'cylinders']].fillna(0)
```

```python
vehicles.loc[condition, 'fuelType']
```

```python
condition2 = vehicles['fuelType'] == 'Regular'
```

```python
avg_displ_subaru = vehicles.groupby(by='make').mean().loc['Subaru', 'displ']
avg_displ_subaru
```

```python
vehicles.loc[condition & condition2, 'displ'] = vehicles.loc[condition & condition2, 'displ'].fillna(avg_displ_subaru)
```

```python
vehicles.loc[condition & condition2, 'displ']
```

```python
vehicles.loc[condition, ['displ', 'cylinders']].fillna(0)
```

```python
vehicles.loc[condition, ['displ', 'cylinders']] = vehicles.loc[condition, ['displ', 'cylinders']].fillna(0)
```

```python
vehicles.displ.isna().sum()
```

# Correct wrong data

```python
# create a condition for which cylinders value == 0
no_cylinder = (vehicles['cylinders'] == 0)


# create a condition for which displacement is different from 0
yes_displacement = (vehicles['displ'] != 0)
```

```python
(no_cylinder & yes_displacement).sum()
```

```python
vehicles.loc[no_cylinder & yes_displacement, 'cylinders'] = 4
```

```python
vehicles.loc[(vehicles['cylinders'] == 0) & (vehicles['displ'] != 0), 'cylinders']
```

# Dropping or checking duplicates rows
    
>    - Dropping fully duplicate row
>    - Subset
>    - `.duplicated()`

```python
df = pd.DataFrame({'cpf':[1,2,2,2,2], 'vlr':[10,20,35,20,25], 'ano':[1992, 1993, 1993, 1993, 1994]})
```

```python
results = (df.groupby('cpf').count() > 3).reset_index()
```

```python
pd.merge(left=df, right=results, on='cpf')
```

```python
df.duplicated(subset=['cpf','ano'], keep=False)
```

```python
df.drop_duplicates(subset=['cpf'])
```

```python
pd.merge(left=df, right=results, on='cpf').drop_duplicates(subset=['cpf','vlr_y',], keep=False )
```

```python
df.sort_values(by=['cpf','ano']).drop_duplicates(subset=['cpf'], keep='last')
```

```python
df.sort_values(by=['cpf','vlr']).drop_duplicates(subset=['cpf'], keep='last')
```

```python
condition1 = df['vlr'] > 20
condition1
```

```python
condition2 = df.duplicated(subset=['cpf'], keep=False)
condition2
```

```python
df.loc[condition1 & condition2]
```
